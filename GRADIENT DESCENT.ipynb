# $$f(x)=x^2+x+1$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1=np.linspace(-3,3,300)
```

```python
def f(x):
    return x**2+x+1
```

```python
def df(x):
    return 2*x+1
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_1,df(x_1))
plt.xlabel('X',fontsize=16,color='g')
plt.ylabel('df(x)',fontsize=16,color='green')

plt.show()
```

# Gradient Descent Algorithm

```python
new_x=3
gamma=0.1

x_list=[new_x]
slope_list=[df(new_x)]

for n in range(100):
    
    previous_x=new_x
    
    new_x=previous_x-gamma*df(previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
```

```python
values=np.array(x_list)
```

```python
f(values)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_1,f(x_1))
plt.scatter(values,f(values),c='r',alpha=0.5,s=100)
plt.xlabel('X',fontsize=16,color='g')
plt.ylabel('f(x)',fontsize=16,color='green')
plt.title('Cost Function')

plt.show()
```

```python
print('local minimum occurs at: ',new_x)
print('df(x) or slope at minimum point is: ',df(new_x))
print('f(x) or cost at this point is: ',f(new_x))
```

# GD-Example:-2

# $$g(x)=x^4-4x^2+5$$

```python
#x_2=np.linspace(start=-2,stop=2,num=1000)

x_2=np.linspace(-2,2,1000)
```

```python
#x_2
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=(20,6))

# 1 Chart: Cost function

plt.subplot(1, 2, 1)


plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()
plt.plot(x_2, g(x_2), color='blue', linewidth=3)

# 2 Chart: Derivative

plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5)


plt.show()
```

```python
def gradient_descent(Initialise,gamma,Derive_func,Iter):
    
    new_x=Initialise
    
    x_list=[new_x]
    slope_list=[Derive_func(new_x)]
    
    for n in range(Iter):
        
        previous_x=new_x
        new_x=previous_x - gamma*(Derive_func(previous_x))
        
        x_list.append(new_x)
        slope_list.append(Derive_func(new_x))
        
    return new_x, x_list, slope_list
```

```python
Local_minima, list_x, list_slope=gradient_descent(Initialise=2,gamma=0.01,Derive_func=dg,Iter=100)
```

### Using Break condition on Ex:-2

```python
def Gradient_Descent(Initialise,gamma,Derive_func,Iter,precision):
    
    new_x=Initialise
    
    x_list=[new_x]
    slope_list=[Derive_func(new_x)]
    
    for n in range(Iter):
        
        previous_x=new_x
        new_x=previous_x - gamma*(Derive_func(previous_x))
        
        step_size=abs(previous_x - new_x)
        
        x_list.append(new_x)
        slope_list.append(Derive_func(new_x))
        
        if step_size < precision:
            break
        
    return new_x, x_list, slope_list, step_size
```

```python
Local_minima, list_x, list_slope, least_stretch=Gradient_Descent(Initialise=2,gamma=0.01,Derive_func=dg,Iter=100,precision=0.001)
```
